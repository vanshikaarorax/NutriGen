```python
def chunk_text(text:str,chunk_size:int=500)->list:
  #splitting text into chunks of fixed 500 size
  chunks=[]
  current_chunk=''
  words=text.split()
  for word in words:
    #checking if adding the word crosses the chunk size
    if len(current_chunk)+len(word)+1<=chunk_size:
      current_chunk+=(word+' ')
    else:
      #store the chunk
      chunks.append(current_chunk.strip())
      current_chunk=word+' '
  if current_chunk:
     chunks.append(current_chunk.strip())
  return chunks

def chunk_pdf_pages(pages_and_texts:list,chunk_size:int=500)->list[dict]:
   #take the pdf text split into text using above function then return page number ,chunk index and text
   all_chunks=[]
   for page in pages_and_texts:
     pages_text=page["text"]
     page_number=page["page_number"]
     chunks=chunk_text(pages_text,chunk_size=chunk_size)
     for i,chunk in enumerate(chunks):
        all_chunks.append({
           "page_number":page_number,
           "chunk_index":i,
           "chunk_char_count":len(chunk),
           "chunk_word_count":len(chunk.split()),
           "chunk_token_count":len(chunk)/4,#1 token=4 chars
           "chunk_text":chunk
       })

   return all_chunks

Fixed_chunked_pages=chunk_pdf_pages(pages_and_texts,chunk_size=500)
print(len(Fixed_chunked_pages))
Fixed_chunked_pages[:2]




```

```python
import random
import textwrap

# Sampling & Pretty Printing
def _scattered_indices(n: int, k: int, jitter_frac: float = 0.08) -> list[int]:
    """
    Evenly spaced anchors + random jitter → indices scattered across [0, n-1].
    """
    if k <= 0:
        return []
    if k == 1:
        return [random.randrange(n)]

    anchors = [int(round(i * (n - 1) / (k - 1))) for i in range(k)]
    out, seen = [], set()
    radius = max(1, int(n * jitter_frac))

    for a in anchors:
        lo, hi = max(0, a - radius), min(n - 1, a + radius)
        j = random.randint(lo, hi)
        if j not in seen:
            out.append(j)
            seen.add(j)

    while len(out) < k:
        r = random.randrange(n)
        if r not in seen:
            out.append(r)
            seen.add(r)

    return sorted(out)


def _draw_boxed_chunk(c: dict, wrap_at: int = 96) -> str:
    """
    Pretty print a single chunk inside a box with metadata.
    """
    header = (
        f" Chunk p{c['page_number']} idx {c['chunk_index']} | "
        f"chars {c['chunk_char_count']} words {c['chunk_word_count']} "
        f"~tokens {int(c['chunk_token_count'])}"
    )

    # Wrap body text nicely
    wrapped_lines = textwrap.wrap(
        c["chunk_text"], width=wrap_at, break_long_words=False, replace_whitespace=False
    )

    content_width = max([0, *map(len, wrapped_lines)])
    box_width = max(len(header), content_width + 2)  # +2 for padding

    top = "╔" + "═" * box_width + "╗"
    hline = "║" + header.ljust(box_width) + "║"
    sep = "╠" + "═" * box_width + "╣"
    body = "\n".join("║ " + line.ljust(box_width - 2) + " ║" for line in wrapped_lines) or \
           ("║ " + "".ljust(box_width - 2) + " ║")
    bottom = "╚" + "═" * box_width + "╝"

    return "\n".join([top, hline, sep, body, bottom])


def show_random_chunks(pages_and_texts: list, chunk_size: int = 500, k: int = 5, seed: int | None = 42):
    """
    Randomly show k scattered chunks from the given pages_and_texts.
    """
    if seed is not None:
        random.seed(seed)

    all_chunks = chunk_pdf_pages(pages_and_texts, chunk_size=chunk_size)
    if not all_chunks:
        print("No chunks to display.")
        return

    idxs = _scattered_indices(len(all_chunks), k)
    print(f"Showing {len(idxs)} scattered random chunks out of {len(all_chunks)} total:\n")

    for i, idx in enumerate(idxs, 1):
        print(f"#{i}")
        print(_draw_boxed_chunk(all_chunks[idx]))
        print()

assert 'pages_and_texts' in globals (), "Run: pages_and_texts = open_and_read_pdf (pdf_path) first."
show_random_chunks (pages_and_texts, chunk_size=500, k=5, seed=42)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import nltk
import numpy as np
from tqdm import tqdm

# Download sentence tokenizer data
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

semantic_model = SentenceTransformer("all-MiniLM-L6-v2")

def semantic_chunk_text(text: str, threshold_value: float = 0.8, max_tokens: int = 500) -> list:
    # Split text into sentences
    sentences = nltk.sent_tokenize(text)
    if not sentences:
        return []

    # Get embeddings for all sentences
    embeddings = semantic_model.encode(sentences)

    chunks = []
    current_chunk = [sentences[0]]
    current_embedding = embeddings[0]

    for i in range(1, len(sentences)):
        token_count = len(" ".join(current_chunk)) // 4
        similarity = cosine_similarity([current_embedding], [embeddings[i]])[0][0]

        if similarity > threshold_value and token_count < max_tokens:
            current_chunk.append(sentences[i])
            current_embedding = np.mean([current_embedding, embeddings[i]], axis=0)
        else:
            chunks.append(" ".join(current_chunk))
            current_chunk = [sentences[i]]
            current_embedding = embeddings[i]

    # Add the last chunk
    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks


def semantic_chunk_pdf(pages_and_texts: list, threshold_value: float = 0.8, max_tokens: int = 500) -> list[dict]:
    """
    Perform semantic chunking across all pages.
    """
    all_chunks = []
    for page in tqdm(pages_and_texts, desc="semantic chunking"):
        pages_text = page["text"]
        page_number = page["page_number"]
        chunks = semantic_chunk_text(pages_text, threshold_value=threshold_value, max_tokens=max_tokens)
        for i, chunk in enumerate(chunks):
            all_chunks.append({
                "page_number": page_number,
                "chunk_index": i,
                "chunk_char_count": len(chunk),
                "chunk_word_count": len(chunk.split()),
                "chunk_text": chunk
            })
    return all_chunks


assert 'pages_and_texts' in globals (), "Run: pages_and_texts = open_and_read_pdf (pdf_path) first."
show_random_chunks (pages_and_texts, chunk_size=500, k=5, seed=42)
semantic_chunked_pages = semantic_chunk_pdf(pages_and_texts)
print(len(semantic_chunked_pages))
print(semantic_chunked_pages[:2])


```

```python
import nltk
nltk.download("punkt", quiet=True)
nltk.download("punkt_tab", quiet=True)

# Manually set NLTK data path
import os
nltk_data_dir = "/root/nltk_data"
if not os.path.exists(nltk_data_dir):
    os.makedirs(nltk_data_dir)
nltk.data.path.append(nltk_data_dir)
```

```python
import nltk
from tqdm.auto import tqdm

nltk.download("punkt")

def recursive_chunk_text(text: str, max_chunk_size: int = 1000, min_chunk_size: int = 100) -> list:
    """
    Recursively splits a block of text into chunks that fit within size constraints.
    Tries splitting by sections, then newlines, then sentences.
    """
    def split_chunk(chunk: str) -> list:
        # Base case
        if len(chunk) <= max_chunk_size:
            return [chunk]

        # Try splitting by double newlines
        sections = chunk.split("\n\n")
        if len(sections) > 1:
            result = []
            for section in sections:
                if section.strip():
                    result.extend(split_chunk(section.strip()))
            return result

        # Try splitting by single newline
        sections = chunk.split("\n")
        if len(sections) > 1:
            result = []
            for section in sections:
                if section.strip():
                    result.extend(split_chunk(section.strip()))
            return result

        # Fallback: split by sentences
        sentences = nltk.sent_tokenize(chunk)
        chunks, current_chunk, current_size = [], [], 0
        for sentence in sentences:
            if current_size + len(sentence) > max_chunk_size:
                if current_chunk:
                    chunks.append(" ".join(current_chunk))
                current_chunk = [sentence]
                current_size = len(sentence)
            else:
                current_chunk.append(sentence)
                current_size += len(sentence)

        if current_chunk:
            chunks.append(" ".join(current_chunk))
        return chunks

    return split_chunk(text)


def recursive_chunk_pdf_pages(pages_and_texts: list,
                              max_chunk_size: int = 1000,
                              min_chunk_size: int = 100) -> list[dict]:
    """
    Takes PDF pages with text and splits them into recursive chunks.
    Returns a list of dicts with page_number, chunk_index, and chunk_text.
    """
    all_chunks = []
    for page in tqdm(pages_and_texts, desc="Recursive chunking pages"):
        page_number = page["page_number"]
        page_text = page["text"]

        chunks = recursive_chunk_text(page_text,
                                      max_chunk_size=max_chunk_size,
                                      min_chunk_size=min_chunk_size)

        for i, chunk in enumerate(chunks):
            all_chunks.append({
                "page_number": page_number,
                "chunk_index": i,
                "chunk_char_count": len(chunk),
                "chunk_word_count": len(chunk.split()),
                "chunk_token_count": len(chunk) / 4,  # rough token estimate
                "chunk_text": chunk
            })

    return all_chunks

recursive_chunked_pages = recursive_chunk_pdf_pages(pages_and_texts)
print(len(recursive_chunked_pages))
print(recursive_chunked_pages[:2])
```

Structural Chunking

```python
import re
import random
import textwrap

# 1) Helper to detect "chapter start" pages
def _is_chapter_header_page(text: str) -> bool:
    # Robust to punctuation/diacritics differences; matches the recurring header
    # e.g., "UNIVERSITY OF HAWAI'I AT MANOA FOOD SCIENCE AND HUMAN NUTRITION PROGRAM"
    return re.search(r"university\s+of\s+hawai", text, flags=re.IGNORECASE) is not None

def guess_title_from_page(text: str) -> str:
    """Best-effort chapter title guess = the text before the 'University of Hawai' header line.
    Falls back to the first ~120 characters."""
    m = re.search(r"university\s+of\s+hawai", text, flags=re.IGNORECASE)
    if m:
        title = text[:m.start()].strip()
        # keep it readable
        title = re.sub(r"\s+", " ", title).strip()
        if 10 <= len(title) <= 180:
            return title
    # fallback
    t = re.sub(r"\s+", " ", text).strip()
    return t[:120] if t else "Untitled Chapter"

# 2) Build chapter chunks with sub-chunking
def chapter_chunk_pdf_pages(pages_and_texts: list[dict],
                            max_chunk_tokens: int = 1000,
                            overlap_tokens: int = 100) -> list[dict]:
    """Returns a list of chapter chunks, split into smaller chunks if needed.

    Args:
        pages_and_texts: List of dicts with 'page_number' and 'text' keys
        max_chunk_tokens: Maximum tokens per chunk (approximate)
        overlap_tokens: Number of tokens to overlap between chunks
    """
    if not pages_and_texts:
        return []

    # Find all page indices that look like the start of a chapter
    chapter_starts = []
    for i, p in enumerate(pages_and_texts):
        txt = p["text"]
        if _is_chapter_header_page(txt):
            chapter_starts.append(i)

    # If nothing detected, treat entire doc as one chapter
    if not chapter_starts:
        chapter_starts = [0]

    # Build chapter ranges and then split into smaller chunks
    all_chunks = []
    chunk_id = 0

    for ci, s in enumerate(chapter_starts):
        # Determine end of this chapter
        e = (chapter_starts[ci + 1] - 1) if (ci + 1 < len(chapter_starts)) else (len(pages_and_texts) - 1)
        if e < s:
            continue  # guard

        pages = pages_and_texts[s:e + 1]
        chapter_text = " ".join(p["text"] for p in pages).strip()
        chapter_title = guess_title_from_page(pages[0]["text"])

        # Estimate tokens (rough: 1 token ≈ 4 chars)
        chapter_tokens = len(chapter_text) / 4

        # If chapter is small enough, keep as one chunk
        if chapter_tokens <= max_chunk_tokens:
            all_chunks.append({
                "chunk_id": chunk_id,
                "chapter_index": ci,
                "title": chapter_title,
                "sub_chunk_index": 0,
                "total_sub_chunks": 1,
                "page_start": pages[0]["page_number"],
                "page_end": pages[-1]["page_number"],
                "chunk_char_count": len(chapter_text),
                "chunk_word_count": len(chapter_text.split()),
                "chunk_token_count": round(len(chapter_text) / 4, 2),
                "chunk_text": chapter_text
            })
            chunk_id += 1
        else:
            # Split chapter into smaller chunks with overlap
            words = chapter_text.split()
            words_per_token = 0.75  # approximate
            words_per_chunk = int(max_chunk_tokens * words_per_token)
            overlap_words = int(overlap_tokens * words_per_token)

            sub_chunks = []
            start_idx = 0
            sub_idx = 0

            while start_idx < len(words):
                end_idx = min(start_idx + words_per_chunk, len(words))
                chunk_words = words[start_idx:end_idx]
                chunk_text = " ".join(chunk_words)

                sub_chunks.append({
                    "chunk_id": chunk_id,
                    "chapter_index": ci,
                    "title": chapter_title,
                    "sub_chunk_index": sub_idx,
                    "page_start": pages[0]["page_number"],
                    "page_end": pages[-1]["page_number"],
                    "chunk_char_count": len(chunk_text),
                    "chunk_word_count": len(chunk_words),
                    "chunk_token_count": round(len(chunk_text) / 4, 2),
                    "chunk_text": chunk_text
                })

                chunk_id += 1
                sub_idx += 1

                # Move to next chunk with overlap
                start_idx = end_idx - overlap_words
                if start_idx >= len(words) - overlap_words:
                    break

            # Update total_sub_chunks for all sub-chunks
            for sc in sub_chunks:
                sc["total_sub_chunks"] = len(sub_chunks)

            all_chunks.extend(sub_chunks)

    return all_chunks


# Usage example
structured_chunked_pages=chapter_chunk_pdf_pages(pages_and_texts, max_chunk_tokens=1000, overlap_tokens=100)


print("=" * 80)
print("First 5 chunks:\n")

for i, chunk in enumerate(structured_chunked_pages[:5]):
    print(f"\n--- CHUNK {i+1} ---")
    print(f"Chunk ID: {chunk['chunk_id']}")
    print(f"Chapter: {chunk['chapter_index']} - {chunk['title']}")
    print(f"Sub-chunk: {chunk['sub_chunk_index'] + 1}/{chunk['total_sub_chunks']}")
    print(f"Pages: {chunk['page_start']}-{chunk['page_end']}")
    print(f"Tokens: {chunk['chunk_token_count']}")
    print(f"Words: {chunk['chunk_word_count']}")
    print(f"Preview: {chunk['chunk_text'][:200]}...")
    print("=" * 80)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Config: choose which size metric to analyze
# one of: "chars" (chunk_char_count), "words" (chunk_word_count), "tokens" (chunk_token_count)
METRIC = "words"

def _size_val(c, metric: str):
    if metric == "chars":
        return c.get("chunk_char_count", len(c.get("chunk_text", "")))
    if metric == "words":
        return c.get("chunk_word_count", len(c.get("chunk_text", "").split()))
    if metric == "tokens":
        # fall back to chars/4 if not present
        return c.get("chunk_token_count", len(c.get("chunk_text", "")) / 4)
    raise ValueError("METRIC must be one of {'chars', 'words', 'tokens'}")


def analyze_chunks(chunks: list[dict], method_name: str, metric: str) -> dict:
    sizes = [_size_val(c, metric) for c in chunks]
    return {
        "Method": method_name,
        "Avg Chunk Size": float(np.mean(sizes)) if sizes else 0.0,
        "Num Chunks": len(sizes),
        "Size Variance": float(np.var(sizes)) if sizes else 0.0,
    }


# Gather results from the previously computed lists
# Make sure your chunked lists exist: chunked_pages, semantic_chunked_pages, recursive_chunked_pages, structure_chunked_pages, llm_chunked_pages
datasets = [
    ("fixed",Fixed_chunked_pages),
    ("semantic", semantic_chunked_pages),
    ("recursive", recursive_chunked_pages),
    ("structure", structured_chunked_pages),
]

results = [analyze_chunks(chks, name, METRIC) for name, chks in datasets]

# Convert results to DataFrame for better display
df = pd.DataFrame(results)
print(df.round(3).to_string(index=False))

# Performance Analysis (summary)
print("\n# Performance analysis")
print("1. Structure-based chunking produced the most coherent sections.")
print("2. Semantic chunking maintained best context preservation.")
print("3. Fixed-size chunking showed lowest variance but poor semantic coherence.")
print("4. Recursive chunking provided balanced results.")
print("5. LLM chunking provided balanced results.")

# Optional: Plot comparison
plt.figure(figsize=(10, 5))
plt.bar(df['Method'], df['Avg Chunk Size'], yerr=np.sqrt(df['Size Variance']), capsize=5)
plt.ylabel(f'Average Chunk Size ({METRIC})')
plt.title('Chunking Methods Comparison')
plt.show()
```

```python
import pandas as pd
import re
import torch
from sentence_transformers import SentenceTransformer, util
from time import perf_counter as timer

# -------------------------------
# 1. Detect device (GPU if available)
# -------------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# -------------------------------
# 2. Load the CSV file
# -------------------------------
df = pd.read_csv("text_chunks_and_embeddings_df.csv")

# -------------------------------
# 3. Convert embeddings column (string -> list of floats)
# -------------------------------
def fix_numpy_array_string(s):
    """
    Fix embeddings saved as NumPy-like strings (no commas).
    Example: "[ 0.12  0.34 -0.56]" -> [0.12, 0.34, -0.56]
    """
    if isinstance(s, str):
        s = s.strip("[]")
        nums = re.split(r"\s+", s.strip())
        try:
            return [float(x) for x in nums if x != ""]
        except ValueError:
            return []
    return s

df["embedding"] = df["embedding"].apply(fix_numpy_array_string)

# -------------------------------
# 4. Convert all embeddings to a tensor and move to device
# -------------------------------
embeddings_tensor = torch.tensor(df["embedding"].tolist(), dtype=torch.float32).to(device)

# -------------------------------
# 5. Load your embedding model on the same device
# -------------------------------
embedding_model = SentenceTransformer("all-mpnet-base-v2", device=device)

# -------------------------------
# 6. Define query and embed it
# -------------------------------
query = "Aging"
print(f"\nQuery: {query}")
query_embedding = embedding_model.encode(query, convert_to_tensor=True).to(device)

# -------------------------------
# 7. Compute similarity using dot product
# -------------------------------
start_time = timer()
dot_scores = util.dot_score(query_embedding, embeddings_tensor)[0]
end_time = timer()
print(f"Time taken: {end_time - start_time:.5f} seconds\n")

# -------------------------------
# 8. Retrieve top-k most similar chunks
# -------------------------------
top_k = 5
top_results = torch.topk(dot_scores, k=top_k)

for score, idx in zip(top_results.values, top_results.indices):
    idx = idx.item()
    print(f"Score: {score.item():.4f}")
    print(f"Text: {df.iloc[idx]['sentence_chunk']}")
    print("=" * 80)
```

```python
from sentence_transformers import util, SentenceTransformer
import torch
from time import perf_counter as timer
import textwrap

# Helper function to print long text nicely
def print_wrapped(text, width=100):
    print(textwrap.fill(text, width=width))
    print("")

# Retrieve function
def retrieve_relevant_resources(
    query: str,
    pages_and_chunks: list[dict],
    model: SentenceTransformer,
    n_resources_to_return: int = 5,
    print_time: bool = True
):
    """Embeds a query with model and returns top k scores and indices from embeddings."""

    # Convert stored embeddings into a tensor
    embeddings_tensor = torch.stack([
        torch.tensor(item["embedding"]) for item in pages_and_chunks
    ])

    # Embed the query
    query_embedding = model.encode(query, convert_to_tensor=True)

    # Compute dot product similarity
    start_time = timer()
    dot_scores = util.dot_score(query_embedding, embeddings_tensor)[0]
    end_time = timer()

    if print_time:
        print(f"[INFO] Time taken to get scores on {len(embeddings_tensor)} embeddings: {end_time - start_time:.5f} seconds.")

    # Ensure k isn’t larger than number of embeddings
    n_resources_to_return = min(n_resources_to_return, len(dot_scores))

    # Top-k results
    scores, indices = torch.topk(dot_scores, k=n_resources_to_return)
    return scores, indices


def print_top_results_and_scores(
    query: str,
    pages_and_chunks: list[dict],
    model: SentenceTransformer,
    n_resources_to_return: int = 5
):
    """Takes a query, retrieves most relevant resources and prints them out in descending order."""

    scores, indices = retrieve_relevant_resources(
        query=query,
        pages_and_chunks=pages_and_chunks,
        model=model,
        n_resources_to_return=n_resources_to_return
    )

    print(f"Query: {query}\n")
    print("Results:")

    # Loop through zipped together scores and indices
    for score, index in zip(scores, indices):
        index = index.item()
        print(f"Score: {score:.4f}")
        print_wrapped(pages_and_chunks[index]["sentence_chunk"])
        print(f"Page number: {pages_and_chunks[index]['page_number']}")
        print("\n")


# ===============================
# Example usage
# ===============================
query = "Aging"

# Get just the scores and indices of top related results
scores, indices = retrieve_relevant_resources(
    query=query,
    pages_and_chunks=pages_and_chunks_over_min_token_len,
    model=embedding_model
)
scores, indices

# Print top relevant sentences
print_top_results_and_scores(
    query=query,
    pages_and_chunks=pages_and_chunks_over_min_token_len,
    model=embedding_model,
    n_resources_to_return=5
)
```

```python


# Helper for printing
def print_wrapped(text, width=100):
    print(textwrap.fill(text, width=width))
    print("")

# --------------------------
# 1️⃣ Prompt Formatter
# --------------------------
def prompt_formatter(query: str, context_items: list[dict]) -> str:
    """
    Formats retrieved context and query into a single RAG-style prompt for GPT.
    """
    context = "- " + "\n- ".join([item["sentence_chunk"] for item in context_items])
    prompt = f"""
You are a helpful AI assistant. 
Based only on the following context items, answer the query as clearly and accurately as possible.
If the context does not provide enough information, say you don’t have sufficient information.

Context:
{context}

User Query: {query}

Answer:
"""
    return prompt.strip()

# --------------------------
# 2️⃣ GPT Generation Function
# --------------------------
def generate_rag_answer(prompt: str, model="gpt-3.5-turbo", max_tokens=400):
    """
    Uses OpenAI model to generate a RAG-augmented answer.
    """
    response = client.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.5,
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content.strip()

# --------------------------
# 3️⃣ Combine Retrieval + Generation
# --------------------------
def get_augmented_answer(
    query: str,
    pages_and_chunks: list[dict],
    embedding_model,
    n_contexts: int = 5
):
    """
    Full RAG pipeline:
    1. Retrieve relevant chunks using sentence-transformers
    2. Format them into a GPT prompt
    3. Generate a context-aware answer
    """

    from sentence_transformers import util
    import torch
    from time import perf_counter as timer

    # Convert stored embeddings to tensor
    embeddings_tensor = torch.stack([
        torch.tensor(item["embedding"]) for item in pages_and_chunks
    ])

    # Embed query
    query_embedding = embedding_model.encode(query, convert_to_tensor=True)

    # Compute dot-product similarity
    start = timer()
    dot_scores = util.dot_score(query_embedding, embeddings_tensor)[0]
    end = timer()
    print(f"[INFO] Retrieval time: {end - start:.4f} sec for {len(embeddings_tensor)} chunks.")

    # Get top-k chunks
    k = min(n_contexts, len(dot_scores))
    scores, indices = torch.topk(dot_scores, k=k)

    context_items = [pages_and_chunks[i.item()] for i in indices]

    # Format prompt
    prompt = prompt_formatter(query, context_items)

    # Generate final answer
    answer = generate_rag_answer(prompt)

    # Print results
    print(f"\n Query: {query}\n")
    print(" Context used:\n")
    for i, c in enumerate(context_items, start=1):
        print(f"{i}. {c['sentence_chunk']}\n")
    print("💬 GPT-Augmented Answer:\n")
    print_wrapped(answer)
    return answer
```
